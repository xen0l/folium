```python
import folium
import folium.plugins as plugins

x , y = -35.19925725627544, 149.0817133320111
dx, dy = 0.001, 0.001
data = [
    [[x+i*dx,y+j*dy] for i in range(5)] for j in range(5)
];

m = folium.Map(location=[-35.19925725627544, 149.0817133320111], zoom_start=15, tiles='Stamen Terrain')
m.add_children(plugins.HeatMapWithTime(data))
m
```
